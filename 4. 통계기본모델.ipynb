## 4. 통계기본모델

### 4.1 통계모델

* 수리모델
* 확률모델
* 통계모델

### 4.2 통계모델을 만드는 방법

* 종속변수 : 변화에 응답하는 변수
* 독립변수 : 대상의 변화를 설명하는 변수 / 다른 대상에 영향을 받지 않는 변수
* 파라메트릭 모델 : 가능 한 현상을 단순화 하여 소수의 파라미터만 사용하는 모델
* 선형모델 : 종속변수와 독립변수의 관계를 선형으로 보는 모델
* 계수와 가중치 : 통계모델에 사용되는 파라미터
* 정보 기준 : 아카이케 정보 기준(AIC) 자주 사용됨. AIC가 작을수록 좋은 모델이라고 판단
* 모델 평가 : 정확도가 높아야 함

### 4.3 데이터의 표현과 모델의 명칭
* 정규선형모델 : 종속변수가 정규분포를 따르는 것을 가정한 선형모델
* 회귀분석 : 정규선형모델 중 독립변수가 연속형 변수인 모델
* 다중회귀분석 : 회귀분성중에 독립변수가 여러게 있는 모델
* 분산분석 : 정규선형모델 중 독립변수가 카데고리형 변수인 모델(일원분산분석, 이원분산분석)
* 일반선형모델
* 머신러닝

### 4.4 파라미터의 추정 : 우도의 최대화
* 우도(likelihood) : 파라미터가 정해져 있을 때 표본을 얻을 수 있는 확률(밀도)을 우도라고 합니다. 
  - 예) 동전을 두번던져서 표본을 얻을 확률( 1/2 * 1/2 = 1/4)
* 우도함수 : 파라미터를 넘겨서 우도를 계산할 수 있는 함수
* 우도로그 : 우도에 로그를 취한 것
* 로그의 성질
  - 지수 : 2^3이라쓰고 2의 3승이라 읽습니다. 
  - 로그 : X의 Y승 = Z의 관계에서 X와Z를 고정해서 Y를 구하는 계산
  - 자연로그 : e : 2.7
  - 성질1 : 정의된 구간에서는 항상 증가
  - 성질2 : 곱셈이 덧셈으로 바뀌게 된다.
  - 성질3 : 절댓값이 극단적으로 작은 값이 되기 어렵다. / 로그를 취하면 컴퓨터로 계산하기 좋은 숫자로 변환
* 최대우도법 : 우도나 로그우도의 결과를 최대로 하기 위한 파라미터를 추정할 때 사용하는 방법
  - 동전던지기 파라미터 1/2, 1/3일때 최대우도는 1/2
* 최대우도추정량(θ) : 햇기호를 사용 θ로 표시
* 정규분포를 따르는 데이터의 우도
* 장애모수 : 직접적인 관심의 대상이 아닌 파라미터
* 정규선형모델의 우도 : stats.norm.pdf

### 4.5 파라미터 추정 : 손실의 최소화
* 손실함수 : 파라미터 추정을 할 때 손실을 최소화하는 목적으로 사용
* 잔차(residuals) : 실제 종속변수의 값과 모델을 이용해서 계산한 종속변수의 추정치와의 차이
* 잔차의 합을 손실함수로 사용하기에는 적합하지 않다.
* 잔차제곱합 : 잔차 제곱해서 구한 것
* 최소제곱법(Ordinary Least Squared) : 잔차제곱합을 최소로 하는 파라미터를 채용하는 방법
* 모집단 분포가 정규분포를 따른때 최소제곱법과 체대우도법의 결과는 일치한다.
* 오차함수 : 머신러닝 분야에서 로그우도를 바꾼 것

### 4.6 예측 정확도의 평가와 변수 선택
* 적합도와 예측 정확도 : 가지고 있는 데이터를 적용했을 때 / 아직 얻지 못한 데이터에 때해 모델을 적용했을때
* 과적합(오버피팅) : 적합도는 높은데 예측 정확도가 낮아지는 경우
* 변수 선택의 의의 : 독립변수를 너무 많이 적용하는경우 과적합을 하게 되는 요인 / 필요없는 독립변수 제외하는것으로 예측도 향상
* 일반화 오차 : 얻지 못한 데이터에 대한 예측오차
* 교차검증 : 데이터를 일정한 규칙에 따라 훈련 데이터와 테스트 데이터로 나누어 예측 정확도를 평가
    - 리브-p-아웃 교차검증 : 가지고 있는 데이터에서 p개의 데이터를 추출하고 남은 데이터를 테스트 데이터로 활용
    - k겹 교차검증 : 가지고 있는 데이터를 k그룹으로 나누고 그 그룹 중 하나를 추출하여 테스트 데이터로 활용
* 아카이케 정보 기준
    - AIC = -2 X (최대로그우도 - 추정된 파라미터수)
    - AIC가 작을수록 좋은모델
* 상대 엔트로피 : 확률분포의 차이를 측정하는 지표
    - 평균로그우도를 최대화하는 것으로 진짜 분포와 예측된 분포와의 차이를 최소화 할 수 있다.
* 평균로그우도가 지니는 편향과 AIC
    - 평균로그우도 자체를 계산하는 것이 어렵기 때문에 최대우도로그를 대신 사용
    - 최대우도로그는 편향된 값을 가지고 있어 편향된 크기만큼(추정된 파라미터 수) 뺌
